# Logistic Regression

```python
from scipy.stats import chi2_contingency
import statsmodels.api as sm
import pandas as pd
```

```python
# Load Data 
df = pd.read_csv('../data/combined_cleaned_data.csv')
df.columns
```

## Build Model

```python
# Step 1: Convert the 'race' column into dummy variables
# We use drop_first=True to avoid the dummy variable trap (over-representation)
df_dummies = pd.get_dummies(df, columns=['race'], drop_first=True)

# Step 2: Define dependent (y) and independent variables (X)
X = df_dummies[['age', 'is_marijuana', 'arrest_district', 'ethnicity', 'home_state', 'sex', 'charge_code']]  # Add other relevant independent variables here
y = df['arrest_count']  # The binary dependent variable

# Step 3: Add a constant (intercept) to the model
X = sm.add_constant(X)

# Step 4: Fit the logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Step 5: Output the results summary
print(result.summary())
```
