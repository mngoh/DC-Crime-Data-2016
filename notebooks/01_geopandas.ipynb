```python
# Load Packages 
import os
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# Load Data 
df = pd.read_csv('../data/combined_cleaned_data.csv')
df.head(2)
```

```python
# Load world map data from Geopandas
# Set the SHAPE_RESTORE_SHX environment variable (in case .shx is missing)
os.environ['SHAPE_RESTORE_SHX'] = 'YES'
world = gpd.read_file('../data/Washington_DC_Boundary.shp')
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['longitude'], df['latitude']))
ax = world.plot(figsize=(10, 10))
gdf.plot(ax=ax, color='red', markersize=5)
# Add title and show the plot
plt.title('Geographic Locations within Washington, DC')
plt.show()
```
