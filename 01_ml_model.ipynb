```python
# Load Packages 
import pandas as pd
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load Data 
dfOrig = pd.read_csv('combined_cleaned_data.csv')
df = dfOrig[['mpd_charge_category', 'mpd_arrest_date', 'charge_code', 'age', 'race', 'home_state', \
         'home_city', 'is_marijuana', 'offense_location_district', 'arrest_count']]
# Rename Columns
df = df.rename(columns = {
    'mpd_charge_category': 'charge_category',
    'mpd_arrest_date':'arrest_date',
    'offense_location_district':'district'
})
```

```python
# Convert Datetime to month and time 
df['arrest_date'] = pd.to_datetime(df['arrest_date'])
df['month'] = df['arrest_date'].dt.month
df['hour'] = df['arrest_date'].dt.hour
```

```python
# Subset the Data 
df = df.drop('arrest_date', axis=1)
df.head(2)
```

## One Hot Encode Features

```python
# One-Hot Encoding using pandas
df = pd.get_dummies(df, columns=['charge_category', 'charge_code', \
                                 'race', 'home_city', 'is_marijuana', 'district'], drop_first=True)
```

```python
df = df.drop(['charge_category', 'charge_code', 'race', 'home_city', 'is_marijuana', 'district'], axis=0)
```

## Kmeans Analysis

```python
# Step 2: Feature Scaling (Standardization)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

# Step 3: KMeans Clustering
# Define the number of clusters (k) -- you can use methods like the Elbow method to find the best k
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Step 4: Add the Cluster Labels to the DataFrame
df['Cluster'] = clusters

# Output the resulting DataFrame
print(df)
```
